```python
#Supporting Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#Keras
from keras.layers import Dense
from keras.utils import to_categorical
from keras.models import Sequential
from keras.callbacks import EarlyStopping
```

```python
df = pd.read_csv('dataset.csv')
df.head()
```

```python
y = to_categorical(df['label'])
X = df.drop('label',axis=1).as_matrix()
X = np.multiply(X, 1/255)
```

```python
#Model
model = Sequential()

#1st layer
model.add(Dense(50, activation='relu', input_shape=(X.shape[1],)))

#2nd layer
model.add(Dense(50, activation='relu'))

#3nd layer
model.add(Dense(50, activation='relu'))

#output layer
model.add(Dense(2, activation='softmax'))

#compile
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

#Training
early_stopping_monitor = EarlyStopping(patience=3)

model.fit(X, y, validation_split=0.2, epochs=30, callbacks=[early_stopping_monitor])
```

```python
model.save('3layers-50nodes.h5')
model.summary()
```
